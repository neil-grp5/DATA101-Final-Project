Data Cleaning and Merging

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

movies = pd.read_csv('movies.csv')
genres = pd.read_csv('genres.csv')
themes = pd.read_csv('themes.csv')
studios = pd.read_csv('studios.csv')
countries = pd.read_csv('countries.csv')



datasets = {'Movies': movies, 'Genres': genres, 'Themes': themes, 'Studios': studios, 'Countries': countries}

for name, df in datasets.items():
    print(f"\n{name} Dataset:")
    print(df.info())


for name, df in datasets.items():
    print(f"\nMissing values in {name} dataset:")
    print(df.isnull().sum())


movies.fillna({'tagline': 'No tagline available', 'description': 'No description available'}, inplace=True)
movies['rating'].fillna(movies['rating'].median(), inplace=True)
movies['date'].fillna(0, inplace=True)
movies['date'] = movies['date'].astype(int)


for name, df in datasets.items():
    print(f"\nDuplicate rows in {name} dataset: {df.duplicated().sum()}")
    df.drop_duplicates(inplace=True)


movies.rename(columns={'id': 'movie_id'}, inplace=True)
genres.rename(columns={'id': 'movie_id'}, inplace=True)
themes.rename(columns={'id': 'movie_id'}, inplace=True)
studios.rename(columns={'id': 'movie_id'}, inplace=True)
countries.rename(columns={'id': 'movie_id'}, inplace=True)



merged_data = movies.copy()

merged_data = merged_data.merge(genres, on='movie_id', how='left')

merged_data = merged_data.merge(themes, on='movie_id', how='left')

merged_data = merged_data.merge(studios, on='movie_id', how='left')

merged_data = merged_data.merge(countries, on='movie_id', how='left')


print("\nMerged Dataset Info:")
print(merged_data.info())


print("\nMissing Values in Merged Dataset:")
print(merged_data.isnull().sum())


plt.figure(figsize=(12,6))
sns.heatmap(merged_data.isnull(), cmap='viridis', cbar=False)
plt.title("Missing Values Heatmap for Merged Dataset")
plt.show()


merged_data.to_csv("merged_movies_dataset.csv", index=False)

print("Data Cleaning & Merging Completed!")
```

Exploratory Data Analysis

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


merged_data = pd.read_csv("merged_movies_dataset.csv")


print("\nDataset Info:")
print(merged_data.info())


print("\nFirst 5 Rows:")
display(merged_data.head())


print("\nMissing Values:")
print(merged_data.isnull().sum())


plt.figure(figsize=(12,6))
sns.heatmap(merged_data.isnull(), cmap='viridis', cbar=False)
plt.title("Missing Values Heatmap")
plt.show()


print("\nDuplicate rows:", merged_data.duplicated().sum())


merged_data.drop_duplicates(inplace=True)


print("\nSummary Statistics:")
display(merged_data.describe())


if 'rating' in merged_data.columns:
    plt.figure(figsize=(8,5))
    sns.histplot(merged_data['rating'].dropna(), bins=20, kde=True)
    plt.title("Distribution of Movie Ratings")
    plt.xlabel("Rating")
    plt.ylabel("Count")
    plt.show()


if 'date' in merged_data.columns:
    plt.figure(figsize=(12,6))
    merged_data['date'].dropna().astype(int).value_counts().sort_index().plot(kind='bar')
    plt.title("Number of Movies Released Per Year")
    plt.xlabel("Year")
    plt.ylabel("Count")
    plt.show()


if 'genre' in merged_data.columns:
    plt.figure(figsize=(12,6))
    merged_data['genre'].value_counts().head(20).plot(kind='bar')
    plt.title("Top 20 Most Common Genres")
    plt.xlabel("Genre")
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.show()


if 'country' in merged_data.columns:
    plt.figure(figsize=(12,6))
    merged_data['country'].value_counts().head(20).plot(kind='bar')
    plt.title("Top 20 Countries by Number of Movies")
    plt.xlabel("Country")
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.show()


numeric_data = merged_data.select_dtypes(include=['number'])


if not numeric_data.empty:
    plt.figure(figsize=(10,6))
    sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
    plt.title("Correlation Matrix for Numeric Columns")
    plt.show()
else:
    print("No numeric columns available for correlation matrix.")


merged_data.to_csv("final_cleaned_dataset.csv", index=False)

print("\n EDA Complete! Cleaned dataset saved as 'final_cleaned_dataset.csv'.")
```

```python
fig = px.choropleth(
    highest_count_genres,
    locations='country',
    locationmode='country names',
    color='genre',
    title='Genre with Highest Count by Country',
    hover_data=['count', 'percentage'],
    color_discrete_sequence=px.colors.qualitative.Set2
)

fig.update_layout(
    coloraxis_colorbar=dict(title='Genre'),
    width=700,
    height=500,
)

fig.show()
```
